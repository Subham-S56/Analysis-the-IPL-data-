```python
# Required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# 
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# bar.plot
plt.figure(figsize=(8,5))
plt.bar(
    list(ipl['player_of_match'].value_counts()[0:5].keys()),
    list(ipl['player_of_match'].value_counts()[0:5]),
    color="g"
)
plt.show()
```
