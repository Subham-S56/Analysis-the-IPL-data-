```python
# Required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# lord match data
ipl = pd.read_csv('matches.csv')
```

```python
# top head of data
ipl.head()
```

```python
# no of row and coulmn
ipl.shape
```

```python
# getting the frequancy of any column
ipl['player_of_match'].value_counts()
```

```python
# getting the top 10 frequancy of any column
ipl['player_of_match'].value_counts()[0:10]
```

```python
# getting the top 5 frequancy of any column
ipl['player_of_match'].value_counts()[0:5]
```

```python
# 
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# bar.plot
plt.figure(figsize=(8,5))
plt.bar(
    list(ipl['player_of_match'].value_counts()[0:5].keys()),
    list(ipl['player_of_match'].value_counts()[0:5]),
    color="g"
)
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
battingfirst= ipl[ipl['win_by_runs']!=0]
```

```python
battingfirst.head()
```

```python
# making a histogram
plt.figure(figsize=(5,7))
plt.hist(battingfirst['win_by_runs'])
plt.title("Distributaion of runs")
plt.xlabel("Runs")
plt.show()
```

```python
battingfirst['winner'].value_counts()
```

```python
# making a bar.plot for top 3 team 
plt.figure(figsize=(6,6))
plt.bar(
    list(battingfirst['winner'].value_counts()[0:3].keys()),
    list(battingfirst['winner'].value_counts()[0:3]),
    color=["blue","yellow","orange"]
)
plt.show()
```

```python
# making a pie chat
plt.figure(figsize=(7, 7))
plt.pie(
    list(battingfirst['winner'].value_counts()),
    labels=list(battingfirst['winner'].value_counts().keys()),
    autopct='%0.1f%%'  
)
plt.show()
```

```python
# batting second reccods
battingsecond= ipl[ipl['win_by_wickets']!=0]
```

```python
battingsecond.head()
```

```python
# making a histogram
plt.figure(figsize=(5,5))
plt.hist(battingsecond['win_by_wickets'],bins=30)
plt.title("Run chasing wins")
plt.xlabel("Runs")
plt.show()
```

```python
battingsecond['winner'].value_counts()
```

```python
# making a bar.plot for top 3 team 
plt.figure(figsize=(7,7))
plt.bar(
    list(battingsecond['winner'].value_counts()[0:3].keys()),
    list(battingsecond['winner'].value_counts()[0:3]),
    color=["purple","blue","red"]
)
plt.show()
```

```python
# making a pie chat
plt.figure(figsize=(7, 7))
plt.pie(
    list(battingsecond['winner'].value_counts()),
    labels=list(battingsecond['winner'].value_counts().keys()),
    autopct='%0.1f%%'  
)
plt.show()
```
